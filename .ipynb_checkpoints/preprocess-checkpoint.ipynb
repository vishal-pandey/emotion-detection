```python
import cv2
import math
```

```python
train_video = "dataset/Train Tom and jerry.mp4"
saveFolder = "save/"
```

```python
cap = cv2.VideoCapture(train_video)
```

```python
frameRate = cap.get(5)
```

```python
i=0
count = 0
```

```python
while(cap.isOpened()):

    cap.set(cv2.CAP_PROP_POS_FRAMES, count)

    count += math.floor(frameRate)
    print(count)

    ret, frame = cap.read()
    # cv2.imshow('frame',frame)



    if (ret != True) or cv2.waitKey(1) & 0xFF == ord('q') or count>=8912:
        break

    cv2.imwrite(saveFolder+"frame"+str(i)+".jpg",frame)
    i+=1
    # if count % math.floor(frameRate) == 0:
```

```python
cap.release()
cv2.destroyAllWindows()
```
